# Making a prediction for a new image.

```python
# Import libraries
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load and prepare the image
def load_image(filename):
    # load the image
    img = load_img(filename, target_size=(224, 224))
    #show the image
    plt.imshow(img)
    # convert to array
    img = img_to_array(img)
    # reshape into a single sample with 3 channels
    img = img.reshape(1, 224, 224, 3)
    # center pixel data
    img = img.astype('float32')
    img = img - [123.68, 116.779, 103.939]
    return img
    
```

```python
# load an image and predict the class
def run_example():
    # load the image
    img = load_image('dataset_validation/cracked/5.jpg')   #Here it is necessary chose the file name that you want to classsify 
    # load model
    model = load_model('crack_classifier_model.h5')
    # predict the class
    result = model.predict(img)
    print('The class probability is:', result[0])
    final_result= np.round_(result[0])
    if final_result==0: #Here 0 is refered to cracked specimens and 1 is refered to uncracked specimens
        print('Cracked specimen')
    else:
        print('Uncraced specimen')
    print(final_result)
    print('Enjoy!!') 
```

```python
# entry point, run the example
run_example()
```
